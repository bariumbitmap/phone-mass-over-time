```python
# import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import math
from adjustText import adjust_text
import numpy as np
```

```python
iphone = pd.read_csv("iphone.tsv", sep='\t')
iphone_pro = pd.read_csv("iphone-pro.tsv", sep='\t')
android = pd.read_csv("nexus_pixel.tsv", sep='\t')
```

```python
android.loc[1, 'Name'] = 'Nexus S\n(AMOLED)'
android.loc[2, 'Name'] = 'Nexus S\n(LCD)'
```

```python
fig, ax = plt.subplots(constrained_layout=True, figsize=(6.4, 2*4.8))
x1 = pd.to_datetime(iphone['Release date'], format='%Y-%m-%d')
y1 = iphone['Weight [g]']
names1 = iphone['Name']
ax.plot(
    x1,
    y1,
    '.-',
    label="iPhone",
    color='lightgray',
    markerfacecolor='black',
    markeredgecolor='black',
)
annotations = []
manual_offset1 = [
    (-300, 1), # iPhone (Gen 1)
    (-400, -2), # iPhone 3G
    (60, -1.5), # iPhone 3GS
    (0, 0), # iPhone 4
    (0, 1), # iPhone 4s
    (-880, -1), # iPhone 5
    (60, -1), # iPhone 5S
    (60, -1), # iPhone 6
    (-660, 1), # iPhone 6s
    (60, -1), # iPhone 7
    (-660, 1), # iPhone 8
    (-1000, 1), # iPhone XR
    (60, -1), # iPhone 11
    (60, -2), # iPhone 12
    (0, 2.5), # iPhone 13
    (0, 1), # iPhone 14
    (-800, -2.5), # iPhone 15
    (60, -1), # iPhone 16
    (60, -1), # iPhone 17
]
for i, (x,y, name, offset) in enumerate(zip(x1, y1, names1, manual_offset1, strict=True)):
    annotations.append(ax.annotate(name, xy=(x+datetime.timedelta(days=offset[0]),y + offset[1]), textcoords='data'))
del i, x, y, name


x3 = pd.to_datetime(iphone_pro['Release date'], format='%Y-%m-%d')
y3 = iphone_pro['Weight [g]']
names3 = iphone_pro['Name']
ax.plot(
    x3,
    y3,
    '.-',
    label="iPhone Pro",
    color='silver',
    markerfacecolor='white',
    markeredgecolor='black',
)
annotations = []
manual_offset3 = [
    (0, -2.5), # iPhone X
    (0, 1), # iPhone XS
    (-1400, -2), # iPhone 11 Pro
    (60, 1), # iPhone 12 Pro
    (-1500, -1.5), # iPhone 13 Pro
    (-1400, 2.5), # iPhone 14 Pro
    (60, 0), # iPhone 15 Pro
    (60, -1), # iPhone 16 Pro
    (-360, 1.5), # iPhone 17 Pro
]
for i, (x,y, name, offset) in enumerate(zip(x3, y3, names3, manual_offset3, strict=True)):
    annotations.append(ax.annotate(name, xy=(x+datetime.timedelta(days=offset[0]),y + offset[1]), textcoords='data'))
del i, x, y, name

x2 = pd.to_datetime(android['Release date'], format='%Y-%m-%d')
y2 = android['Weight [g]']
names2 = android['Name']
ax.plot(
    x2,
    y2,
    '.-',
    label="Google flagship",
    color='lightgreen',
    markerfacecolor='green',
    markeredgecolor='green',
)
manual_offset2 = [
    (-1100, -1), # Nexus One
    (-200, -5), # Nexus S (AMOLED)
    (-360, 1), # Nexus S (LCD)
    (700, -10), # Galaxy Nexus
    (0, 0), # Nexus 4
    (0, 0), # Nexus 5
    (60, -1), # Nexus 5X
    (0, -3), # Pixel
    (60, -1), # Pixel 2
    (60, -1), # Pixel 3
    (-360, 1), # Pixel 4Galaxy Nexus
    (60, -1), # Pixel 5
    (-700, -1), # Pixel 6
    (-60, 1), # Pixel 7
    (0, -3), # Pixel 8
    (0, -2.5), # Pixel 9
    (60, -0.5), # Pixel 10
]
for i, (x, y, name, offset) in enumerate(zip(x2, y2, names2, manual_offset2, strict=True)):
    annotations.append(ax.annotate(name, xy=(x + datetime.timedelta(days=offset[0]),y + offset[1]), textcoords='data', color='green'))
galaxy_nexus = android.loc[android['Name'] == 'Galaxy Nexus'].iloc[0]
galaxy_nexus_x = pd.to_datetime(galaxy_nexus['Release date'], format='%Y-%m-%d')
galaxy_nexus_y = galaxy_nexus['Weight [g]']
ax.annotate(
    "",
    xytext=(galaxy_nexus_x + datetime.timedelta(days=manual_offset2[3][0]), galaxy_nexus_y + manual_offset2[3][1]),
    xy=(galaxy_nexus_x + datetime.timedelta(days=100), galaxy_nexus_y-2),
    arrowprops=dict(arrowstyle="->", color="green")
)
del i, x, y, name

# adjust_text(annotations);
ax.axhline(95, color="black", linestyle="--", zorder=0)
ax.annotate("Motorola Razr V3 (2004)", xy=(datetime.datetime(year=2020, month=1, day=1), 97), color="black")
ax.axhline(133, color="blue", linestyle="--", zorder=0)
ax.annotate("Nokia 3310 (2000)", xy=(datetime.datetime(year=2021, month=1, day=1), 135), color="blue")
ax.axhline(226.8, color="lightgray", linestyle="--", zorder=0);
ax.annotate("Roll of US quarters", xy=(datetime.datetime(year=2021, month=1, day=1), 223), color="dimgray")
ax.spines[['right', 'top']].set_visible(False)
ax.legend(loc='upper left', bbox_to_anchor=(0., 0.7))
ax.set_xlabel("Release date")
ax.set_ylabel("Weight [g]")
years = mdates.YearLocator(2) # every two years
ax.xaxis.set_major_locator(years)
ax.set_xlim(right=datetime.datetime(year=2027, month=12, day=31))
# ax.set_ylim(bottom=0)
fig.autofmt_xdate()
ax.set_title("Flagship smartphones have gotten heavier")
pass;
```
